```python
#성별, 무게 최소
print(df.groupby('Gender')['Weight'].mean())
```

```python
#그룹화하기 DataFrameGroupBy 객체 반환
grouped_smoke_df = df.groupby('SMOKE')
print(type(grouped_smoke_df))
```

```python
print(grouped_smoke_df)
```

```python
#그룹화하기 비만척도별, 성별, 키,무게 최소값 출력
multi_group_var = df.groupby(['0be1dad','Gender'])[['Weight', 'Height']].mean()
print(multi_group_var)
```

```python
#데이터 프레임 평탄화
flat = multi_group_var.reset_index()
print(flat)
```

```python
#중복 제외 개수 비만척도별, 교통수단개수
print(df.groupby('0be1dad')['MTRANS'].nunique())
```

```python
#지정한 열이나 행의 개수 구함
print(df.groupby('0be1dad')['Gender'].value_counts())
```

```python
#데이터를 그래프로 표현하기
global_age_weight = df.groupby('0be1dad')[['Age','Weight']].mean()
print(global_age_weight)
```
